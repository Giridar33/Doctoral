```python

df_symptoms = pd.read_csv("dataset.csv")
df_diets = pd.read_csv("diets.csv")
df_medications = pd.read_csv("medications.csv")
df_symptom_description = pd.read_csv("symptom_Description.csv")
df_symptom_precaution = pd.read_csv("symptom_precaution.csv")
df_symptom_severity = pd.read_csv("Symptom-severity.csv")
```

```python
def clean_text(df):
    df.columns = df.columns.str.lower().str.strip()
    return df.applymap(lambda x: x.lower().strip() if isinstance(x, str) else x)
```

```python
df_symptoms = clean_text(df_symptoms)
df_diets = clean_text(df_diets)
df_medications = clean_text(df_medications)
df_symptom_description = clean_text(df_symptom_description)
df_symptom_precaution = clean_text(df_symptom_precaution)
df_symptom_severity = clean_text(df_symptom_severity)
```

```python
# Ensure 'disease' column exists in all datasets
def ensure_column(df, column_name):
    if column_name not in df.columns:
        print(f"Warning: '{column_name}' column not found in dataset.")
    return df
```

```python
df_diets = ensure_column(df_diets, "disease")
df_medications = ensure_column(df_medications, "disease")
df_symptom_description = ensure_column(df_symptom_description, "disease")
df_symptom_precaution = ensure_column(df_symptom_precaution, "disease")
df_symptoms = ensure_column(df_symptoms, "disease")
```

```python
# Merge disease-related datasets
try:
    df_disease_info = df_diets.merge(df_medications, on="disease", how="outer")
    df_disease_info = df_disease_info.merge(df_symptom_description, on="disease", how="outer")
    df_disease_info = df_disease_info.merge(df_symptom_precaution, on="disease", how="outer")
except KeyError as e:
    print(f"KeyError: {e}. Check if 'disease' column exists in all datasets.")
    print("df_diets columns:", df_diets.columns)
    print("df_medications columns:", df_medications.columns)
    print("df_symptom_description columns:", df_symptom_description.columns)
    print("df_symptom_precaution columns:", df_symptom_precaution.columns)
    exit()
```

```python
# Ensure 'disease' column exists before merging
if "disease" in df_symptoms.columns and "disease" in df_disease_info.columns:
    df_merged = df_symptoms.merge(df_disease_info, on="disease", how="left")
else:
    print("Error: 'disease' column not found in one of the datasets.")
    print("df_symptoms columns:", df_symptoms.columns)
    print("df_disease_info columns:", df_disease_info.columns)
    exit()

# Scale symptoms based on severity
for symptom in df_symptom_severity["symptom"].values:
    if symptom in df_merged.columns:
        weight = df_symptom_severity[df_symptom_severity["symptom"] == symptom]["weight"].values[0]
        df_merged[symptom] *= weight  # Scale symptom presence by severity
```

```python
# Save the final processed dataset
final_merged_file_path = "final_merged_dataset.csv"
df_merged.to_csv(final_merged_file_path, index=False)

print("Final dataset saved successfully as", final_merged_file_path)
```
